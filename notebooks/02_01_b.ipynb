# Simple linear regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from pylab import rcParams
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 10, 8
```

```python
# Create a random number of rooms
rooms = 2*np.random.randn(100,1)+3
rooms[:10]
```

```python
# Create random house prices
price = 265 + 6*rooms + abs(np.random.randn(100,1))
price[:10]
```

```python
# Plot rooms against price
plt.plot(rooms, price, 'r^')
plt.xlabel('Num of Rooms, 2024 Average')
plt.ylabel('2024 Average home price, 1000s USD')
plt.show()
```

```python
x = rooms
y = price
# Least squares linear regression of number of rooms against house price
Linreg = LinearRegression()
Linreg.fit(x,y)
print(Linreg.intercept_, Linreg.coef_)
```

*Simple Algebra*
- y = mx + b
- b = intercept = 265.68

*Estimated Coefficients*
- LinReg.coef_ = [6.01] Estimated coefficients for the terms in the linear regression problem.

```python

print(Linreg.score(x,y))
```
